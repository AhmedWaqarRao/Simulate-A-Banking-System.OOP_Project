```python
class Bank:
    
    def __init__(self):
        self._mydict={}
        
    def updatedatabase(self):
        self._mydict[self.userid]=[self.name,self.amount]
        print(self._mydict)
        
    def verifywithdrawal(self,requestedamount):
        if (self.userid in self._mydict) and (self.name==self._mydict[self.userid][0]):
            if(requestedamount<=self._mydict[self.userid][1]):
                print('Valid Transaction. Balance Available')
                self._mydict[self.userid][1]-=requestedamount
                print('Withdrawral Successful.')
                print('\n')
                print('User Id: ',self.userid)
                print('Remaining Balance: ',self._mydict[self.userid][1])
                print('\n')
            else:
                print('Withdrawal amount exceeds available balance')
                print('\n')
        else:
            print('Invalid Account. Enter a valid name and Account Number.')
            print('\n')
    
    def verifydeposit(self,depositamount):
        if (self.userid in self._mydict) and (self.name==self._mydict[self.userid][0]):
            print('Valid Account.')
            self._mydict[self.userid][1]+=depositamount
            print('Deposit Successful.')
            print('\n')
            print('User Id: ', self.userid)
            print('New Balance after Deposit: ',self._mydict[self.userid][1])
            print('\n')
        else:
            print('Invalid Account. Enter a valid name and account number to deposit amount.')
            print('\n')
```

```python
import random
user=User()

while True:
    print('Enter 1 to create a new account')
    print('Enter 2 to access existing account')
    print('Enter 3 to display your available balance')
    print('Enter 4 to exit')
    print('\n')
    
    userchoice=int(input('Enter your choice: '))
    
    print('\n')
    
    if userchoice is 1:
        user.newaccount()
        user.updatedatabase()
        
    elif userchoice is 2:
        print('Welcome. \n')
        user.existingaccount()
        print('D:Deposit')
        print('W:Withdraw')
        print('\n')
        decision=input('Enter your choice: ')
        
        if decision.lower()=='d':
            depositamount=int(input('Enter your amount to be deposited: '))
            user.verifydeposit(depositamount)
            
        elif decision.lower()=='w':
            requestedamount=int(input('Enter your amount you want to withdraw: '))
            user.verifywithdrawal(requestedamount)
            
    elif userchoice is 3:
        user.displaybalance()
            
    elif userchoice is 4:
        print('Thanks for using our banking system.')
        break
```
